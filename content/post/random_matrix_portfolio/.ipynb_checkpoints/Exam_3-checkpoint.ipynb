## Sign Prediction

```python
def insertLogSign(df):
    """ This functions inserts the sign of return, sign of momentum into the dataframe.
    Args
    ----
        df: dataframe
    Return
    ----
        df: dataframe

    """
    df['y'] = np.where(df['Adj Close'].shift(-1) > df['Adj Close'],1,-1)
    df['Previous Sign'] = np.where(df['Adj Close'] > df['Adj Close'].shift(1),1,-1)
    return df
```

## Generate the features to be used in the model

```python
def insertOpenCloseDiff(df):
    """ This function inserts the open-close difference into the dataframe.

    Args
    ----
        df : dataframe

    Return
    ------
        df : dataframe

    """
    df['O-C'] = df['Open'] - df['Close']
    return df

def insertHighLowDiff(df):
    """ This function inserts the HighDiff column into the dataframe.

    Args
    ----
        df : dataframe

    Return
    ------
        df : dataframe

    """
    df['H-L'] = df['High'] - df['Low']
    return df

def insertSimpleMovingAverage(df, n):
    """ This function inserts the Simple Moving Average into the dataframe for the given n.

    Args
    ----
        df : dataframe
        n : int
    Return
    ------
        df : dataframe
    
    """
    df['SMA' + str(n)] = df['Adj Close'].rolling(n).mean()
    return df


def insertExponentialMovingAverage(df, n):
    """ This function inserts the Exponential Moving Average into the dataframe for the given n.

    Args
    ----
        df : dataframe
        n : int
    Return
    ------
        df : dataframe
    """
    df['EMA' + str(n)] = df['Adj Close'].ewm(span=n).mean()
    return df


def insertMomentum(df, n):
    """ This function inserts the Momentum into the dataframe for the given n.

    Args
    ----
        df : dataframe
        n : int
    Return
    ------
        df : dataframe
    """
    df['Momentum' + str(n)] = df['Adj Close'] - df['Adj Close'].shift(n)
    return df

def insertReturn(df, n):
    """ This function inserts the Return into the dataframe for the given n.

    Args
    ----
        df : dataframe
        n : int
    Return
    ------
        df : dataframe
    """
    # check if has Return n column exist
    columnName = f'Return{n}'
    df[columnName] = np.log1p((df['Adj Close'] - df['Adj Close'].shift(n))/df['Adj Close'].shift(n)) 
    return df

def filterDataFrame(df, start_date, end_date):
    """
    Args
    ----
        df : dataframe
        start_date : str
        end_date : str

    Return
    ------
        df : dataframe


    """
    df = df[start_date:end_date]
    # as requested in CQF exam,  we will categorize the dataframe using the smalest returns
    # we will drop the rows that have returns less than 10e-6
    insertReturn(df, 1)
    # fill NaN values in Return1 column with 1
    df['Return1'].fillna(1, inplace=True)
    df = df[np.abs(df['Return1']) > 10e-6]
    # we don't need Return1 column anymore
    df.drop(columns=['Return1'], inplace=True) 
    return df
```

## Prediction Task

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
# parameters
test_size = 0.2
seed_number = 42
```

```python
def filterDfNaN(df):
    """ This function will remove the rows with NaN values.
    In addition will also remove the last row of the dataframe.
    Args
    ----
        df : dataframe
    """
    df.dropna(inplace=True)
    df = df[:-1]
    return df
```

### Question 2 (a) - produce a table comparing L1 and L2 type of penalisation: the impact made on regression coefficients; plotting is optional here;

```python
df = filterDataFrame(dfSP500, '2016-01-01', '2021-01-01')

insertLogSign(df)
insertSimpleMovingAverage(df, 7)
insertExponentialMovingAverage(df, 3)
insertOpenCloseDiff(df)
insertHighLowDiff(df)

for n in range(2, 5, 1):

    insertReturn(df, n)
    insertMomentum(df, n)

filterDfNaN(df)
y = df.pop('y')
cols = set(df.columns) - set(['Low', 'High', 'Volume', 'Open', 'Close', 'Adj Close', 'Previous Sign'])
X = df[cols]

print(f"Features {X.columns}")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=seed_number)
X_train = (X_train - X_train.mean()) / X_train.std()
X_test = (X_test - X_test.mean()) / X_test.std()

```

```python
# set numpy print options maximum digits
try:
    from mlxtend.evaluate import bias_variance_decomp
except importError:
    print("mlxtend not installed: !pip install mlxtend")

np.set_printoptions(precision=3)
df_coefs = []
coefs_intercept = []
decomp_list = []
models = {'l1':None, 'l2':None}
for penalty in ['l1', 'l2']:
    solver = 'saga'
    print(f'Penalty: {penalty}, solver: {solver}')
    print('-'*30)
    clf = LogisticRegression(
        penalty=penalty, solver=solver, class_weight='balanced', random_state=42).fit(X_train, y_train)
    models[penalty] = clf
    print(f"Accuracy Train {clf.score(X_train, y_train)*100:.2f}%")
    print(f"Accuracy Test {clf.score(X_test, y_test)*100:.2f}%")
    clf.predict(X_test).mean()

    # create a dataframe with the coefficients and feature names
    col_name = f'Coefs for penalty {penalty}  and solver {solver} '
    coefs = pd.DataFrame(np.abs(clf.coef_[0]), index=X.columns, columns=[col_name])
    mse, bias, var = bias_variance_decomp(
        clf, X_train.values, y_train.values, X_test.values, y_test.values, 
        loss='mse', num_rounds=500, random_seed=42)
    decomp_list += [(mse, bias, var)]
    # sort the coefficients by their absolute value
    coefs.sort_values(col_name, ascending=False, inplace=True)
    df_coefs += [coefs]

    print(f'MSE: {mse:.3f}, Bias: {bias:.3f}, Variance: {var:.3f}')
    print('\n')
```

### Question 2 (b) -  explain in bold font whether L1 or L2 regression likely to have a high bias and low variance.

```python
df = filterDataFrame(dfSP500, '2016-01-01', '2021-01-01')
insertLogSign(df)
insertSimpleMovingAverage(df, 7)
insertExponentialMovingAverage(df, 3)
insertOpenCloseDiff(df)
insertHighLowDiff(df)

for n in range(2, 5, 1):
    insertReturn(df, n)
    insertMomentum(df, n)

filterDfNaN(df)
y = df.pop('y')
cols = set(df.columns) - set(['Low', 'High', 'Volume', 'Open', 'Close', 'Adj Close', 'Previous Sign'])
X = df[cols]


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=test_size, random_state=seed_number)
X_train = (X_train - X_train.mean()) / X_train.std()
X_test = (X_test - X_test.mean()) / X_test.std()


decomp_list = []
regParameters = np.linspace(0.001, 60, 50) 
for regParameter in regParameters:
    C = 1 / (regParameter)
    solver = 'liblinear'
    penalty = 'l1'
    clf = LogisticRegression(
        penalty=penalty, 
        solver=solver, 
        class_weight='balanced', 
        max_iter=4000,
        random_state=42, C=C).fit(X_train, y_train)
   
    clf.predict(X_test).mean()
    # create a dataframe with the coefficients and feature names
   
    mse, bias, var = bias_variance_decomp(
        clf, X_train.values, y_train.values, X_test.values, y_test.values, 
        loss='mse', num_rounds=100, random_seed=42)
    decomp_list += [[mse, bias, var]]
    # sort the coefficients by their absolute value

    # show the coeficients of the  features in the model
    #coefs = pd.Series(clf.coef_[0], index=X.columns)
    #print(coefs.sort_values())
decomp_list = np.array(decomp_list)
```

### Bias-Variance and the relatioship with the regularization parameter

```python
# plot the bias-variance decomposition evolution for different regularization parameters
decomp_list = np.array(decomp_list)
plt.figure(figsize=(10, 5))
fig, ax = plt.subplots(1, 1)
#plt.plot(regParameters, decomp_list[:, 0], label='MSE')
ax.plot(regParameters, decomp_list[:, 1], label='Bias', c='tomato')
axr = ax.twinx()
axr.plot(regParameters, decomp_list[:, 2], label='Variance', c='steelblue')
plt.legend()
plt.xlabel('Regularization Parameter')
ax.set_ylabel('Bias')
axr.set_ylabel('Variance')
plt.title('Bias-Variance Decomposition Evolution')
plt.show()
```

 ### Question 2 (b) - Conclusion:
 Model
* Ridge-l2: 
    * Squared value of coefficients (${\beta^2}$)
    * Low parameter regularization: Original Logistic Regression
    * High parameter regularization: Underfitting
    * Effect: Reduces the importance of some features.[1,2]
    * Great for: When number of features is greater than the number of observation [1]. When there is correlated features; collinearity; dependency.
    * Bad for: Feature selection[1].
* Lasso-l1: 
    * Absolute value of the coefficients (${|\beta|}$)
    * Low parameter regularization: Original Logistic Regression
    * High parameter regularization: Underfitting
    * Effect: Reduces the importance of some features, forcing some coefficients to zero.[1,2]
    * Great for: Feature selection.
    * Bad for: When there is collinearity

* Conclusion: 
    * Both regressions (Ridge & Lasso) aim to avoid overfitting, consequently reducing the variance and increasing the bias.
    * The results of the mlxented package shows the tradeoff between reducing the variance and increasing the bias.
    * In the experiment, L1 has presented a slightely higher bias than L2, and lower variance than L2.

Reference:
* [1] Zou, H & Hastie, T 2005, 'Regularization and variable selection via the elastic net', _Journal of the royal statistical society: Series B (statistical methodology)_, vol. 67, no. 2, pp. 301-320.
* [2] Dietterich, T. G, & Kong, E.B 1995, 'Machine learning bias, statistical bias, and statistical variance of decision tree algorithms', _Technical report_, Department of Computer Science, Oregon State University.

### Question 3 (a) - choose at least two approaches and briefly indicate the main property and key maths for each.For example, VIF focuses on interdependent (colinear) features.

### Shapley values

Shapley Values is a method from coalitional game theory. The method assumes that each "player" in the game is a feature where the prediction is the payout. Therefore,the method calculates how to best distribute the payout among the features. Thus, the shapley value is defined for each instance of the data; 
[Shapley](https://shap.readthedocs.io/en/latest/index.html)

The contribution of the feature $i$, for a  prediction $y$ is given by 
$$
Shapley_j(y) = \beta_j x_j - \mathbb E (\beta_j \mathbf X_j).
$$



### Variance Inflation Factor (VIF)
VIF is a method which detects Multicollinearity. The calculation will measure the correlation between the features. A high VIF indicates strong multicollinearity between features. Therefore the method recommendeds to drop the features which have a VIF between 5-10

## Part B. Mathmatics of Supervised Learning

### Question 6. Briefly present the maths of the logistic classifier by writing down: use β for coefficients not θ
#### Question 6 (a) - the complete multivariate cost function;


Let $n$ be the number of features and $m$ the number of samples.

We have the coeficients 
$$
\mathbf \beta= \begin{bmatrix} \beta_0 \\ \beta_1 \\ \vdots \\ \beta_n \\ \end{bmatrix}. 
$$
The vector that represents a sample is given by
$$
\mathbf x = \begin{bmatrix} 1 \\ x_1 \\ \vdots \\ x_n \\ \end{bmatrix}.
$$
We can expresss the estimated outcome as 
$$
\hat y^{(i)} = \frac{1}{1+e^{\beta^T \mathbf x^{(i)}}},
$$
where $x^{(i)}$ is the $i$-th sample.

Using the matricial notation described above, the cost function is 
$$
\mathrm{Cost}(\hat y^{(i)}, y) = -\log(\hat y^{(i)}) y - \log(1-\hat y^{(i)})(1- y),
$$
where $y=0, 1$

#### Question 6 (b) - log-loss form of the MLE log-likelihood function;

$$
\hat y^{(i)} = \frac{1}{1+e^{\beta^T \mathbf x^{(i)}}}
$$
$$ 
J(\beta) = 
-\frac{1}{m}
\sum\limits^{m}_{i=1}
    \left(
        y_{i} \log(\hat y^{(i)})
        + (1-y_{i}) \log (1 - \hat y^{(i)})
    \right), 
$$

### Question 7. Consider MSE (β̂) wrt to the true value β in context of regression methods,

$$ E[(\hat{\beta} - \beta)^{2}] = Var[\hat{\beta}] + (E[\hat{\beta}] - \beta])^{2}$$

#### (a) can there exist an estimator with the smaller MSE than minimal least squares?

Answer -> No, there is no estimator with a smaller MSE that can be computed using just the data (Accordingly to the Gauss-Markov Theorem).

#### (b) for a prediction, does the MSE measure an irreducible error or model error?

Answer -> No, the MSE is the model error which has the irreducible error ($Var[\hat{\beta}]$), as a lower bound.
